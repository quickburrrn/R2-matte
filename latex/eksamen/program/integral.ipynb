### Oppgave 4.b)

Dette programmet regner summen med midpunkttiln√¶rming/rektangelmetoden

```python
start = 0
slutt = 2
n = 100

dx = (slutt - start)/n

def f(x):
    return 3**(2*x)

def bedre_metode():
    areal = 0

    for i in range(1, n+1):
       areal += f(start + (i - 1/2) * dx) * dx

    return areal

print(f"arealet er: {bedre_metode()}")
```
