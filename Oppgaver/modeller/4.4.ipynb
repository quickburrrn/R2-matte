En elekstriker måler resistans i like lange metalledene med ulike diameter. Hun fører resultatene inn i tabellen.

```python
from pylab import *
import matplotlib.pyplot as plt


# Data og plotting for å vurdere passelig modell

diameter = [0.19, 0.40, 0.70, 1.10, 1.40]
resistans = [3.30, 0.75, 0.24, 0.099, 0.061]

plt.scatter(diameter, resistans, label="Måledata")

plt.plot(diameter, resistans, "--", label="Sammenheng")

plt.xlabel("Diamaeter (mm)")
plt.ylabel("Resistant (Ω)")
plt.title("Resistans vs Diameter")

plt.grid(True)
plt.legend()

plt.show()
```

```python
# a lag en modell

from pylab import *
from scipy.optimize import curve_fit

def pwer(mm, a, b):
    return a * mm**b

[a, b] = curve_fit(pwer, diameter, resistans)[0]

print("a =", round(a, 2))
print("b =", round(b, 2))

plot(diameter, resistans, "o")
xlabel("Diameter (mm)")
ylabel("Resistans (Ω)")
grid(True)
mm = linspace(0, 2, 100)
plot(mm, pwer(mm, a, b), "r")
```
