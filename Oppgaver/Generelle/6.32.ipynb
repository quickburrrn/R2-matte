# finn $\int_0^\pi sin x dx$

Med rektangelmetoden med 50 rektangler.

```python
# bruker venstretilnærming siden tilnærmingsmetode ikke er spesifisert

import numpy

a = 0           # fra
b = numpy.pi    # til
n = 50          # res

S = 0
d = numpy.pi / n

for i in range(n):
    S += numpy.sin(a + d*i) * d

print(S)
```
