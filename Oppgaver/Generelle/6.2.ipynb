# Lag et programm for
$a_1=1$ og $a_{n+1} = a_n + n$

```python
def a(n):
    if n == 1:
        return 1
    else:
        return a(n-1) + n

for n in range(1, 51):
    print(f"{n}", a(n))

# mest sansyblig har fasit feil her
```

Endre programmet slik at det summerer de 50 f√∏rste leddene

```python
def a(n):
    if n == 1:
        return 1
    else:
        return a(n-1) + n

for n in range(1, 50):
    print(f"{n+1}", a(n+1)-a(n))
```

### c)

Endre programmet i a slik at det summerer de 50 leddene, hva er summen?

```python
def a(n):
    if n == 1:
        return 1
    else:
        return a(n-1) + n

s = 0

for n in range(1, 51):
    s += a(n)

print(f"summen er : {s}")
```
