Hvor lavt må du sette eulermetoden for å kunne se at tilnærmingen blir forskjellig fra den analytiske løsningen?

modellen vi fant i eksempel 9 er: $N=10071,429 e^{-0.007t}-3571.429$ med en differensialligning på : $N'=-0,007 \cdot N -25$

```python
from pylab import *

a = 0 # start
b = 50 # slutt
n = 5 # antall steg må sette den til ca 5 for å kunne se forskjellen fra den analytiske løsningen
delta_t = (b - a)/n # steglengde

N0 = 6500 # N initialbetegnelse

def Nder(N):
    return -0.007*N - 25

t = zeros(n)
N = zeros(n)

N[0] = N0

for i in range(n - 1):
    N[i+1] = N[i] + Nder(N[i]) * delta_t
    t[i+1] = t[i] + delta_t

plot(t, N)
plot(t, 10071.429 * np.exp(-0.007*t) - 3571.429)
xlabel("Tid (år)")
ylabel("Innbyggertall")
show()
```
