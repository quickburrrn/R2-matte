$f(x)=6,2-9,8sin(0,017x + 1,35), x \in [1, 365]$

### a)

Tegn grafen til $f$

````python
from pylab import *

# definerer planet
x = linspace(-1, 365, 100)

# definerer funksjonen
y = 6.2-9.8*sin(0.017x + 1.35)

# plott
plot(x, y, label="f(x) = x^2")

# legget til pynt
xlabel("x")
ylabel("y")
title("Plot av f(x) = x^2")
legend()
grid(True)

show()

print("hello world")
````

### b)

Finn likevektslinja, amplituden og perioden. Hva forteller svarene om normaltemperaturen?

````python
# vi har f(x)=6,2-9,8sin(0,017x + 1,35), x in [1, 365]

likevektslinja = 6.2 # denne finner vi med √• sette sin(x) = 0 eller ser p√• konstantleddet

amplitude = 9.8 # denne finner vi bare med √• se p√• funksjonen

# perioden er definert med 2pi/c hvor c for oss er 0,017
perioden = (2*pi/0.017)

print(likevektslinja)
print(amplitude)
print(perioden)
````
