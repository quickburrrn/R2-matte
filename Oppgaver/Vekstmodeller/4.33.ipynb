En harepopulasjonhar i dag 50 individer, og vi antar at den følger en logistisk vekstmodell der bæreevnen er 500 individer og proporsjonalitetskonstanten er 0,06. Tiden måler vi i måneder.

Bruk eulermetoden for å beskrive utviklingen i harepopulasjonen i årene som kommer.

```python
from pylab import *

a = 0 #start
b = 200 # slutt
n = 10000 # steg
delta_t = (b - a)/n

N0 = 50 # initialbetegnelsen

def Nder(N):
    return 0.06 * N * (1-N/500) # definerer differensialigning

t = zeros(n)
N = zeros(n)

N[0] = N0 # setter initalbetegnelse

for i in range(n - 1):
    N[i+1] = N[i] + Nder(N[i]) * delta_t
    t[i+1] = t[i] + delta_t

plot(t, N)
xlabel("Tid (år)")
ylabel("Innbyggertall")
show()
```
