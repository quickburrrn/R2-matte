En isfisker havner i en råk og kroppstemperaturen synker raskt.

Kropstemperaturen $T(t)^\circ C$ etter $t$ minutter er gitt ved differensialikningen

$$T'=-0.011 \cdot T$$

Gjør nødvengie antakelser og tegn en kurve som viser hvordan kroppstemperaturen utvikler seg.

```python
# Vi må anta initialbetegnelsen, jeg antar han starter med kroppstemperatur på 37 grader som er vanlig
from pylab import *

a = 0 # start
b = 60 * 10 # slutt (etter en time)
n = 10000 # antall steg
delta_t = (b - a)/n

T0 = 37 # vi antar at han starter 37

def Tder(T):
    return -0.011 * T

t = zeros(n)
T = zeros(n)

T[0] = T0

for i in range(n - 1):
    T[i+1] = T[i] + Tder(T[i]) * delta_t
    t[i+1] = t[i] + delta_t

plot(t, T)
xlabel("Tid (minutter)")
ylabel("Temperatur Celcius")
show()
```
