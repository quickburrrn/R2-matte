En vanntank inneholder $200 L$ rent vann. EN saltoppløsning tilsettes så med konstant fart, samtidig som det tappes ut like mye væske i bunnen av vanntanken. Vanntanken inneholder $y(t) g$ salt etter $t$ minutter, der $y(t)$ er gitt: $$y'=10-0,01y$$

a)

Bruk eulermetoden til å lage en grafisk framstilling av saltmengden i vanntankende første 8 timene

```python
from pylab import *

a = 0 # start
b = 8*60 # slutt 8 timer

n = 10000 # antall steg (oppløsning)
delta_t = (b - a)/n # legnde mellom hvert steg

N0 = 0 # intialbetegnelsen (antall salt i starten)

def Nder(N):
    return 10 - 0.01 * N

t = zeros(n)
N = zeros(n)

N[0] = N0

for i in range(n - 1):
    N[i+1] = N[i] + Nder(N[i]) * delta_t
    t[i+1] = t[i] + delta_t

plot(t, N)
xlabel("Tid (minutter)")
ylabel("Salt (g)")
show()
```

Hvor mye salt kan det maksimalt bli i tanken?

```python
from pylab import *

# jeg løser denne opppgaven med å finne et punkt lang tid vekke

a = 0 # start
b = 8*6000 # slutt 8 timer

n = 1000000 # antall steg (oppløsning)
delta_t = (b - a)/n # legnde mellom hvert steg

N0 = 0 # intialbetegnelsen (antall salt i starten)

def Nder(N):
    return 10 - 0.01 * N

t = zeros(n)
N = zeros(n)

N[0] = N0

for i in range(n - 1):
    N[i+1] = N[i] + Nder(N[i]) * delta_t
    t[i+1] = t[i] + delta_t

print(N[len(N)-1], "er et punkt langt vekke")
print("Det virker som tanken maks kan ta", round(N[len(N)-1]), "gram salt")
```
