# Rettetester

Tanken her er at man skal kunne teste en rekursiv eller eksplisitt formel/samenheng.

Dette er egent hvis man har noen definerte tall, skal lage en formel så verifsere formelen

```python
# Definerte tall består av tuple(verdi, n)
definert = [
    (2, 1),
    (6, 2),
    (12, 3),
    (20, 4),
    (30, 5),
]

# bytt logikk til den du skal teste

# def rekursiv(n):
#     if n == 1:
#         return 1**3 #a_1
#     else:
#         return rekursiv(n-1) + n**3 # funksjon

def eksplisitt(n):
    return n**2 + n

feilet = False

# tester
for i in definert:
    resultat = eksplisitt(i[1]) # bytt til hva det skal være
    if(i[0] == resultat):
        print(f"Riktig: n={i[1]}, var {resultat} skal være: {i[0]}")
    else:
        print(f"Feilet: n={i[1]} var {resultat} skulle være være: {i[0]}")
        feilet = True

print()
print('feilet' if feilet else 'suksess')
```
