```
# Importer det du trenger
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

sp.init_printing()
```

## Regning / utledning
Bruk både håndregning (forklar med tekst) og Python til å sjekke mellomresultater.

### 1. Definer variable og funksjoner

```
# Eksempel: bytt ut med det som passer i oppgaven
x = sp.symbols('x', real=True)
f = x**3 - 3*x + 1  # eksempel
f
```

### 2. Beregninger (derivasjon, likningsløsning, osv.)

```
# Deriver
f_deriv = sp.diff(f, x)
f_deriv
```

```
# Løs likninger
kritiske_punkt = sp.solve(sp.Eq(f_deriv, 0), x)
kritiske_punkt
```

### 3. Grafisk sjekk

```
x_vals = np.linspace(-5, 5, 400)
f_vals = [f.subs(x, xv) for xv in x_vals]

plt.figure()
plt.axhline(0)
plt.axvline(0)
plt.plot(x_vals, f_vals)
plt.title('Graf til f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
