Husk å skrive *både* matematisk løsning og tolkning i ord.

```
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

sp.init_printing()
```

## 1. Sett opp problemet

```
# Eksempel: y' = k*y, y(0) = y0
t = sp.symbols('t', real=True)
y = sp.Function('y')(t)
k, y0 = sp.symbols('k y0', real=True)

diff_eq = sp.Eq(sp.diff(y, t), k*y)
diff_eq
```

## 2. Løs differensiallikningen symbolsk

```
generell_losning = sp.dsolve(diff_eq)
generell_losning
```

## 3. Bruk startbetingelse

```
# Eksempel med y(0) = y0
C1 = sp.symbols('C1')
los_expr = generell_losning.rhs
betingelse = sp.Eq(los_expr.subs(t, 0), y0)
C1_los = sp.solve(betingelse, sp.symbols('C1'))
C1_los
```

```
los_med_betingelse = sp.simplify(los_expr.subs(sp.symbols('C1'), C1_los[0]))
los_med_betingelse
```

## 4. Tolkning og graf

```
# Sett inn konkrete verdier fra oppgaven, f.eks. k=0.2, y0=100
k_val = 0.2
y0_val = 100
y_t = sp.lambdify(t, los_med_betingelse.subs({k: k_val, y0: y0_val}), 'numpy')

t_vals = np.linspace(0, 10, 200)
y_vals = y_t(t_vals)

plt.figure()
plt.plot(t_vals, y_vals)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Løsning av differensiallikningen')
plt.grid(True)
plt.show()
```
