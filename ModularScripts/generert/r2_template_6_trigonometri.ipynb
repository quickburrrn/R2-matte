```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

sp.init_printing()
```

## 1. Definer variabel og funksjon

```python
# Eksempel: f(x) = 3*sin(2x - pi/3) + 1
x = sp.symbols('x', real=True)
f = 3*sp.sin(2*x - sp.pi/3) + 1
f
```

## 2. Finn amplitude, periode og faseforskyvning

```python
# For funksjon på formen f(x) = a*sin(bx + c) + d
a = 3
b = 2
c = -sp.pi/3
d = 1

amplitude = abs(a)
periode = 2*sp.pi/abs(b)
faseforskyvning = -c/b  # husk minus!
vertikal_forskyvning = d
amplitude, periode.simplify(), faseforskyvning.simplify(), vertikal_forskyvning
```

## 4. Grafisk sjekk av trig-funksjon

```python
x_vals = np.linspace(0, 2*np.pi, 400)
f_num = sp.lambdify(x, f, 'numpy')
f_vals = f_num(x_vals)

plt.figure()
plt.axhline(0)
plt.axvline(0)
plt.plot(x_vals, f_vals)
plt.title('Trigonometrisk funksjon')
plt.xlabel('x')
plt.grid(True)
plt.show()
```
