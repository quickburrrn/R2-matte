```python
import sympy as sp
import numpy as np

sp.init_printing()
```

## 1. Definer følge / rekke

```python
# Eksempler – erstatt med din egen følge
n = sp.symbols('n', integer=True, nonnegative=True)
a_n = 2*n + 1        # eksplisitt formel
a_n
```

## 2. Første ledd og noen eksempler

```python
[a_n.subs(n, k) for k in range(6)]  # a_0, a_1, ..., a_5
```

## 3. Sjekk om følge/rekke er aritmetisk eller geometrisk

```python
a = [a_n.subs(n, k) for k in range(5)]
differanser = [a[k+1] - a[k] for k in range(4)]
forhold = [sp.simplify(a[k+1]/a[k]) for k in range(4) if a[k] != 0]
differanser, forhold
```

## 4. Sum av de n første leddene

```python
S_n = sp.summation(a_n, (n, 0, sp.symbols('N', integer=True, nonnegative=True)))
S_n
```

## 5. Uendelig geometrisk rekke

```python
# Hvis du har førsteledd a_1 og kvotient k, kan du sjekke konvergens:
a_1, k = sp.symbols('a_1 k', real=True)
S_uendelig = sp.simplify(a_1/(1-k))
S_uendelig
```
